```python
def tier_1(col):
    if '\n' in col:
        s = col.split('\n')
        col = ' '.join(s)
    
    s = col.split()
    col = ' '.join(s)
    
    print(col)
    if ("indian institute of technology" in col.lower()):
        print('IITs')
        return 'IITs'
    elif ("national institute of technology" in col.lower()):
        return 'NITs'
    elif ("jawaharlal nehru centre" in col.lower()):
        return 'JNCASR' 
    elif ("indian institute of science education and research" in col.lower()):
        return 'IISER'
    elif ("560012" in col.lower()):
        return 'IISc'
    
    elif ("laboratory" in col.lower()):   # NEERI not included, CDRI included
        return 'CSIR Labs'
    elif ("center for" in col.lower()):   # NEERI not included, CDRI included
        return 'CSIR Labs'
    elif ("central" in col.lower()):
        if ("central university" in col.lower()):
            return np.nan
        else:
            return 'CSIR Labs'
    
    elif ("national institute of science education" in col.lower()):
        return 'Other TIER-I'
    elif ("tata institute of" in col.lower()):  #TIFR
        return 'Other TIER-I'
    elif ("indian association" in col.lower()):  # IACS
        return 'Other TIER-I'
    elif ("national institute for interdisciplinary" in col.lower()):  # NIIIST, Thiruvanthapuram
        return 'Other TIER-I'
    
    
    elif ("birla" in col.lower()):
        return 'BITS'
    elif ("vit university" in col.lower()):
        return 'VIT'
    elif ("vellore institute of technology" in col.lower()):
        return 'VIT'
    
    else:
        return np.nan
```

```python
CRG_18_19['TIER-I'] = np.nan
CRG_18_19['TIER-I'] = CRG_18_19['PI Details'].apply(tier_1)
```

```python
PI_Name = []
Insti = []
Pincode = []
States = []
for k,x in enumerate(CRG_18_19['PI Details']):
    lst = x.split(',')
    PI_Name.append(lst[0])
    Insti.append(lst[1])
#     Pincode.append(lst[-1])
    Pincode.append(x[-6:])
    state = lst[-2]
    if '\n' in state:
        s = state.split('\n')
        state = ' '.join(s)
    
    s = state.split()
    state = ' '.join(s)
    
    States.append(state)
```

```python
States
```

```python
CRG_18_19['PI Name'] = PI_Name
CRG_18_19['Institute'] = Insti
CRG_18_19['State']=States
CRG_18_19['Pincode']=Pincode
```
